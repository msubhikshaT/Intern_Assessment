## * **Import Libraries** *

```python
from tqdm.notebook import tqdm
import tensorflow as tf
import tensorflow_hub as hub
import pandas as pd
```

##* **Load the Universal Sentence Encoder module** *

```python
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

## * **Function to get USE embeddings for a given text** *

```python
def get_use_embeddings(text):
    embeddings = embed([text])[0].numpy()
    return embeddings
```

## * **Loading the CSV Data** *

```python
file_path = "/content/d1.csv"

# Read the CSV file using pandas
df = pd.read_csv(file_path)
```

## * **Generate Embeddings for Data** *

```python
# Assuming df is your DataFrame
text_column = df.iloc[:, 0]  # Select all rows from the first column (index 0)

# Get USE embeddings for each text in the column
for line in tqdm(text_column, desc="Getting USE embeddings"):
    embeddings = get_use_embeddings(str(line))
    print(f"Line: {line}\nEmbeddings: {embeddings}\n")
```
